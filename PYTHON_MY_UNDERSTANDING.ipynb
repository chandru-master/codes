
# Python program to find n-th stair 
# using step size 1 or 2 or 3.
 
# Returns count of ways to reach n-th
# stair using 1 or 2 or 3 steps.

```python
def findStep(n) :
    if (n == 1 or n == 0) :
        return 1
    elif (n == 2) :
        return 2
     
    else :
        return findStep(n - 3) + findStep(n - 2) + findStep(n - 1)
 
 
# Driver code
n = int(input("enter the number of steps : "))
print(findStep(n))
```

```python
# Python program to count
# ways to reach nth stair

# Recursive function to find
# Nth fibonacci number
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)



# Returns no. of ways to
# reach sth stair
def countWays(s):
    return fib(s + 1)

# Driver program
s = 4
print("Number of ways = ",countWays(s))


# Contributed by Harshit Agrawal
```

## Binomial Coefficient | DP-9

```python
def binomialCoeff(c, k):
 
    if k > c:
        return 0
    if k == 0 or k == c:
        return 1
 
    # Recursive Call
    return binomialCoeff(c-1, k-1) + binomialCoeff(c-1, k)
 
c = 5
k = 2
print("Value of C(%d,%d) is (%d)" % (c, k, binomialCoeff(c, k)))
```

# 8 queen sequence

```python
from itertools import permutations

N=8
sol=0
cols = range(N)
for combo in permutations(cols):                      
    if N==len(set(combo[i]+i for i in cols))==len(set(combo[i]-i for i in cols)):
        sol += 1
        print('Solution '+str(sol)+': '+str(combo)+'\n')  
        print("\n".join(' o ' * i + ' X ' + ' o ' * (N-i-1) for i in combo) + "\n\n\n\n")
```

# init method which means initialisation

```python
class time:
    """represents the time and attributes: hour, minute,seconds"""
    
    def __init__(self,hour = 0,minute = 0,seconds = 0):
        self.hour = hour
        self.minute = minute
        self.seconds = seconds
    def print_time(t):
        print("hour" ,t.hour,"\nMinutes", t.minute,"\nseconds",t.seconds)
        
        
# need to create the object of the class:
print("----------------------")
time1 = time()
time1.print_time()

print("----------------------")
time2 = time(10,20,11)
time2.print_time()
```
